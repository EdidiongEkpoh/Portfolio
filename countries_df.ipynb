Project Summary: Creates and edits pandas dataframe about the countries in the world

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('countries_of_the_world.csv', index_col=0, decimal=',')
```

```python
df
```

Formats the dataframe so that there is no leading and trailing
whitespace in the "Region" column and row labels, also capitalizes Regions normally

```python
df['Region'] = df['Region'].str.strip().str.title()
df.index = df.index.str.strip()
df
```

Creating a "Growth Rate" column in the dataframe

```python
df['Growth Rate'] = df['Birthrate'] - df['Deathrate']
```

Helper functions that takes a current population of a country along with its death rate
in order to find the year that the country will be extinct

```python
def dod(p, r):
  num_yrs = 0
  while p > 2:
    p = p + p * r / 1000
    num_yrs += 1
  return num_yrs
```

Creates a 'Years to Extinction' column that will be filled with
values if the rows growth rate is negative. Uses dod helper function
to determine number of years until extinction

```python
df['Years to Extinction'] = np.nan
for row in df.index:
  if df.loc[row,'Growth Rate'] < 0:
    df.loc[row, 'Years to Extinction'] = dod(df.loc[row, 'Population'], df.loc[row, 'Growth Rate'])
```

Creates a series of the 'Years to Extinction' column with no na values

```python
negative = df['Years to Extinction']
negative = negative.dropna()
series = negative.sort_values()
```

Prints the 5 countries that are closest to extinction

```python
count = 0
for row, value in series.items():
    print(str(row).strip() + ': ' + str(value) + ' Years to Extinction')
    count += 1
    if count == 5:
      break
```
